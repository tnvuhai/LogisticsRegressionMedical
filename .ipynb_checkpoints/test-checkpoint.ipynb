```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_excel("DataSet.xlsx")
dataset.head(10)
```

```python
corr = dataset.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
# input
x = dataset.iloc[:, [1,2,3,6]].values

# output
y = dataset.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train) 
X_test = sc_x.transform(X_test)
  
print (X_train[0:10, :])
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 40)
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = classifier.predict(X_test)
qd = confusion_matrix(y_test, y_pred)
print ("Confusion Matrix : \n", qd)
```

```python
classifier.score(X_train, y_train)
```

```python
classifier.score(X_test, y_test)
```

```python
#Tranform model
classifier.predict(X_test)
```

```python
import math
def PredictPro(gos, training, credit, experience, intercept):
    gosPara = classifier.coef_[0][0]
    trainingPara = classifier.coef_[0][1]
    creditPara = classifier.coef_[0][2]
    experiencePara = classifier.coef_[0][3]
    logarit = intercept + gosPara*gos - trainingPara*training + \
    creditPara*credit + experiencePara*experience
    Decision = 1/(1+math.exp(-logarit))
    
    if(Decision >= 0.5):
        return 1
    else:
        return 0
    
print(PredictPro(0,1,0,12,  classifier.intercept_))
```
